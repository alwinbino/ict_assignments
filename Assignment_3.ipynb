```python
def biggest(a,b,c):
  if a>b and a>c:
    print(a,'is the biggest')
  elif b>a and b>c:
    print(b,'is the biggest')
  else:
    print(c,'is the biggest')
```

## *2*

```python
def sum_list(items):
    sum_numbers = 0
    for x in items:
        sum_numbers += x
    return sum_numbers

l = [1, 2, 3, 4, 5]
print(sum_list(l))
```
