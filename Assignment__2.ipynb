```python
for number in range(1, 101):
    if number % 3 == 0 and number % 5 == 0:
        print("FizzBuzz")
    elif number % 3 == 0:
        print("Fizz")
    elif number % 5 == 0:
        print("Buzz")
    else:
        print(number)
```

# ***Q2***

```python
u = int(input("Enter the number of units consumed: "))
t = 0
if u <= 100:
    t = u * 5
elif u <= 200:
    t = (100 * 5) + ((u - 100) * 7)
else:
    t = (100 * 5) + (100 * 7) + ((u - 200) * 10)
print(f"total electricity bill is:", (t))
```
