```python
import math

a = float(input("Enter the length of side a: "))
b = float(input("Enter the length of side b: "))
c = float(input("Enter the length of side c: "))
s = (a + b + c) / 2
area = math.sqrt(s * (s - a) * (s - b) * (s - c))
print(f"The area of the triangle is: {area:.2f}")
```

```python
wall_length = float(input('Enter the length of the wall (in meters): '))
wall_height = float(input('Enter the height of the wall (in meters): '))
painting_rate = 25
window_length = 1.5
window_height = 1.2
total_wall_area = wall_length * wall_height
one_window_area = window_length * window_height
two_windows_area = 2 * one_window_area
area_to_be_painted = total_wall_area - two_windows_area
total_cost = area_to_be_painted * painting_rate
print(f"Total wall area: {total_wall_area:.2f} sq meters")
print(f"Total window area (2 windows): {two_windows_area:.2f} sq meters")
print(f"Total area to be painted: {area_to_be_painted:.2f} sq meters")
print(f"Total cost to paint the wall: ₹{total_cost:.2f}")
```
