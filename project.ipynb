# **2**. **Student Performance Analyzer**

```python
students = []
def add_student():
    roll = int(input("Enter roll number: "))
    name = input("Enter name: ")
    marks = {}
    for subject in ["Math", "Science", "English"]:
        while True:
            m = int(input(f"Enter marks for {subject}: "))
            if 0 <= m <= 100:
                marks[subject] = m
                break
            else:
                print("Marks must be between 0 and 100")
    total = 0
    for m in marks.values():
        total += m
    average = total / 3
    result = "Pass"
    for subject in marks:
        if marks[subject] < 40:
            result = "Fail"
    student = {
        "roll": roll,
        "name": name,
        "marks": marks,
        "average": average,
        "result": result
    }
    students.append(student)
    print("Student Mark added successfully!\n")
def display_students():
    if len(students) == 0:
        print("No student records found.\n")
        return
    for s in students:
        print("\nRoll:", s["roll"])
        print("Name:", s["name"])
        print("Marks:")
        for sub in s["marks"]:
            print(sub, ":", s["marks"][sub])
        print("Average:", s["average"])
        print("Result:", s["result"])
while True:
    print("\n--- Student Performance Analyzer ---")
    print("1. Add Student Mark")
    print("2. Display Students")
    print("3. Exit")
    choice = input("Enter your choice: ")
    if choice == "1":
        add_student()
    elif choice == "2":
        display_students()
    elif choice == "3":
        print("Program Ended")
        break
    else:
        print("Invalid choice! Try again.")
```

# **9. Text-Based Decision Game**


```python
score = 0
choices = []

print("WELCOME TO THE ADVENTURE GAME")
print("Your journey begins...\n")
print("You are standing at a crossroad.")
print("Do you go LEFT or RIGHT?")
choice1 = input("Enter your choice: ").lower()
if choice1 == "left":
    print("You find a small village.")
    score = score + 10
    choices.append("Went Left")
else:
    print("You enter a dark forest.")
    choices.append("Went Right")
print("\nYou see a river.")
print("Do you SWIM across or WALK along it?")
choice2 = input("Enter your choice: ").lower()
if choice2 == "swim":
    print("You cross the river safely.")
    score = score + 10
    choices.append("Swam Across")
else:
    print("You get tired while walking.")
    score = score - 5
    choices.append("Walked Along River")
print("\nNight is coming.")
print("Do you CAMP or KEEP WALKING?")
choice3 = input("Enter your choice: ").lower()
if choice3 == "camp":
    print("You rest and regain energy.")
    score = score + 10
    choices.append("Camped")
else:
    print("You collapse from exhaustion.")
    score = score - 10
    choices.append("Kept Walking")
print("\nGAME OVER")
print("Your Score:", score)
if score >= 25:
    print("Ending: You win and reach home safely!")
elif score >= 10:
    print("Ending: You survive but are still lost.")
else:
    print("Ending: You lose the adventure.")
print("\nYour Choices:")
for c in choices:
    print("-", c)
play_again = input("\nDo you want to play again? (yes/no): ")
if play_again == "yes":
    print("Restart the program to play again.")
else:
    print("Thank you for playing!")
```
